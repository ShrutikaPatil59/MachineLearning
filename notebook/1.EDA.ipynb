```python
import pandas as pd

# Now load the dataset
df = pd.read_csv("data/stud.csv")
```

```python
# -------------------------
# 1. Basic info
# -------------------------
print("First 5 rows:")
print(df.head(), "\n")

print("Dataset info:")
print(df.info(), "\n")

print("Statistical summary for numeric columns:")
print(df.describe(), "\n")

print("Checking for missing values:")
print(df.isnull().sum(), "\n")
```

```python

# -------------------------
# 2. Distribution of categorical columns
# -------------------------
categorical_cols = ["gender", "race_ethnicity", "parental_level_of_education", "lunch", "test_preparation_course"]

for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=df, palette="Set2")
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python

# -------------------------
# 3. Distribution of scores
# -------------------------
score_cols = ["math_score", "reading_score", "writing_score"]

for col in score_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], bins=15, kde=True, color="skyblue")
    plt.title(f"Distribution of {col}")
    plt.show()

# Boxplots to check outliers
for col in score_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(y=df[col], color="lightgreen")
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
# -------------------------
# 4. Correlation analysis
# -------------------------
plt.figure(figsize=(8,6))
corr = df[score_cols].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation between scores")
plt.show()
```

```python
# -------------------------
# 5. Scores vs categorical features
# -------------------------
for cat in categorical_cols:
    for score in score_cols:
        plt.figure(figsize=(6,4))
        sns.boxplot(x=cat, y=score, data=df, palette="pastel")
        plt.title(f"{score} vs {cat}")
        plt.xticks(rotation=45)
        plt.show()
```

```python
# Numeric columns (scores)
score_cols = ["math_score", "reading_score", "writing_score"]
```

```python
# Plot histograms
for col in score_cols:
    plt.figure(figsize=(6,4))
    plt.hist(df[col], bins=15, color='skyblue', edgecolor='black')
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()

# -------------------------
# Using Seaborn for nicer plots
# -------------------------
for col in score_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], bins=15, kde=True, color='lightgreen')
    plt.title(f"Distribution of {col} (with KDE)")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```
